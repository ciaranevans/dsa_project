# Choose cell to view

```python
cell_no = 7001
label_fp = "/Users/ciaran/data/imagery/tifs/blue_anchor/train/labels/sand/blue_anchor_" + str(cell_no) + ".tif"
beach_fp = "/Users/ciaran/data/imagery/tifs/blue_anchor/train/images/blue_anchor_" + str(cell_no) + ".tif"
label_image = rasterio.open(label_fp)
beach_image = rasterio.open(beach_fp)
```
